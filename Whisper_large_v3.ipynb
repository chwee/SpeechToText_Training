```python
import whisper

model_size = "large-v3"

model = whisper.load_model(model_size)
result = model.transcribe("sample4.mp3")
print(result["text"])
```

```python
result = model.transcribe("sample4.mp3")
print(result["text"])
```

```python
segments = result["segments"]

for segment in segments:
    print("[%.2fs -> %.2fs] %s" % (segment['start'], segment['end'], segment['text']))
```
